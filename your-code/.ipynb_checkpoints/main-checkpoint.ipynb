```python


# Your code
df=pd.DataFrame()
for i in range(1,10):
    url=f'https://rickandmortyapi.com/api/character?page={i}'
    response = requests.get(url)
    results = response.json()
    flattened_data = pd.json_normalize(results['results'])
    d = pd.DataFrame(flattened_data)
    df = pd.concat([df,d])
df
```

## Bonus

Extract the following information from the data:

* The character who appeared in most episodes (aside from Rick and Morty, of course!).

* The total number of dead characters.

* A list of all character origins (without duplicates).

```python
# your code here
url = 'https://rickandmortyapi.com/api/character'
res = requests.get(url)
res_json = res.json()
flattened_data = pd.json_normalize(res_json['results'])
df = pd.DataFrame(flattened_data)
```

```python
df['episode'].str.len()
```

```python
df['length'] = df['episode'].str.len()
df.sort_values(by=['length'],ascending = False)
```

```python
tabla_final = df.sort_values(by=['length'],ascending = False)[['name','image','length']]
```

```python
df=pd.DataFrame()
for i in range(0,42):
    url=f'https://rickandmortyapi.com/api/character?page={i}'
    response = requests.get(url)
    results = response.json()
    flattened_data = pd.json_normalize(results['results'])
    d = pd.DataFrame(flattened_data)
    df = pd.concat([df,d])
df
```
